```python
def beforeHeavenBaGua():  
    seq = ['乾', '兑', '离', '震', '巽', '坎', '艮', '坤'];  
  
    for i in range(8):  
        for j in range(8):  
            print('{0} + {1} = {2}, '.format(seq[i], seq[j], seq[(i+j)%8]), sep = ', ', end = '');  
            print('{0} - {1} = {2}, '.format(seq[i], seq[j], seq[(i-j)%8]), sep = ', ', end = '');  
            print('{0} * {1} = {2}'.format(seq[i], seq[j], seq[(i*j)%8]), sep = ', ', end = '');  
  
            print(' ');

beforeHeavenBaGua()
```
